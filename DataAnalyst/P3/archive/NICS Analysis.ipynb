That's better!  Now, what was that data?

```python
census_notes = census_notes.reset_index(drop=True)
census_notes
```

Interesting.  It looks like these are footnotes, presumably to reduce the data size?  Let's try replacing Fact Notes.

```python
notes = census_notes.iloc[5:8][['Fact', 'Fact Note']]

notes = notes.set_index('Fact', drop=True)
notes
```

```python
notes_map = notes.to_dict()['Fact Note']
```

```python
notes_map
```

```python
census_data['Fact Note'] = census_data['Fact Note'].map(notes_map)
```

```python
census_data
```

That's better.  But state was a **field** in the gun data.  Better transpose the census_data to make it easier to join.

```python
census_data.set_index('Fact', inplace=True)
census_data = census_data.transpose()
```

```python
census_data
```

That's better, although all that work converting the Fact Notes will be less than useful since it won't carry over.  Also, for now we're disregarding the Value Flags
